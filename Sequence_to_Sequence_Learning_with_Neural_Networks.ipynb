# Seq2Seq Learning_with_Neural_Networks 논문 리뷰

- Seq2Seq의 의미는 하나의 시퀀스를 다른 시퀀스로 번역을 하겠다. 라는 의미를 가진다.
  
  
- Seq2Seq 이후에 Attention이 나왔고 이를 이용하여 만든것이 Transformer이다.
    - 그 이후에 Transformer를 이용해 만든 모델들이 GPT, BERT와 같은 모델이다.

### Seq2Seq 의 구조

- 인코더 -> context vector(고정된 크기) -> 디코더 방식



### 언어모델이란?
- 문장(시퀀스)에 확률을 부여하는 모델을 의미한다.
- 특정 상황에서 적절한 단어나 문장을 예측할 수 있다.

## 통계적 언어 모델
- 통계적 언어모델은 카운트 기반의 접근을 사용
- 하지만 현실적으로 너무 방대한 양의 데이터가 필요하고 처음보는 문장에 대한 확률이 0이 된다.
- 그러한 해결책으로 나온것이 N-gram 언어 모델이다.
    - 이는 인접한 일부 단어만 고려하자는 아이디어의 모델이다.

## RNN 기반의 seq2seq
- 결과적으로 통계적 언어 모델보다 훨씬 좋은 성능을 보여준다.
- 해결하기 위한 문제로 문맥 벡터를 추출하도록 한다.
- LSTM을 이용하여 문맥 벡터를 추출하는데 여기서 특징은 마지막 hidden_state만 사용하는 것이다.
    - 이는 마지막 hidden_state는 모든 문장 정보를 담고 있기 때문이다.
- 특징으론 인코더와 디코더가 서로 다른 파라미터(가중치)를 지닌다.



## Seq2Seq 성능 개선 포인트
### LSTM 활용 및 입력 문장의 순서 뒤집기

- 특이점으로 출력 문장의 순서는 바꾸지 않으나 입력문장의 순서를 거꾸로 했을 때 더 높은 정확도를 보이는 점이 있다.
  
  
- 이는 일반적인 언어의 특성으로 앞쪽에 위치한 단어들 기리 연관성이 높고 거의 주어가 먼저 등장하기 때문이다.

# 논문 리뷰

- BLEU score를 이용하여 기계번역 스코어를 매겼다.
- 이는 가장 대중적인 평가지표이다.

- LSTM을 4개의 층으로 쌓아 구현하였다
- 입력문장의 순서를 뒤집어서 사용하였더니 잘 되었음.

- spacy 라는 라이브러리를 사용
    - 자연어처리에서 많이 사용되는 라이브러리로 문장의 토큰화 , 태깅 등의 전처리 기능을 위한 라이브러리 ** 공부할것

- 7.5epochs만큼 학습시켰고 0.5 epoch마다 learning_rate를 절반으로 줄였다.
- batch_size는 128
- 패딩은 가장 긴 문장을 기준으로 패딩하였음.
